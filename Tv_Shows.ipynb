# Introduction

### I will be working with the TV shows on various streaming platforms dataset downloaded from kaggle. The dataset comprises a comprehensive list of tv shows available on these streaming platforms. 

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data Loading

```python
url=r"C:\Users\Sowjanya\tv_shows.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.rename(columns= {'Unnamed: 0':'ID'}, inplace = True)
df.drop(['type'], axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```


# Data Cleaning

### Replacing all the Nan Values


```python
#Age Column
df['Age'].mode()
```

```python
df['Age'].replace(np.nan,'16+',inplace=True)
```

```python
df['Age'].replace('all','1+',inplace=True)
df.tail(20)
```

```python
#IMDb Column
df['IMDb'].min()
```

```python

df['IMDb'].replace(np.nan,'1',inplace=True)
df.info()
```

```python
#Rotten Tomatoes Column
df['Rotten Tomatoes']
df['Rotten Tomatoes'].replace(np.nan,'0',inplace=True)
df.isnull().sum()


```

### Changing the Data Types


```python
df.dtypes
```

```python
df['IMDb']=df['IMDb'].apply(pd.to_numeric)
df['Rotten Tomatoes'] = df['Rotten Tomatoes'].str.rstrip('%').astype('float')/10.0
#df['Age'] = df['Age'].str.replace('+','').astype('float')
df.dtypes
```

### How did the release of number of shows increased over the years?

```python
releases_2000 = df[df.Year >= 2000]
plat_df=releases_2000.groupby('Year')[['Netflix','Prime Video','Hulu','Disney+']].sum()
plat_df
```

```python

plt.plot(plat_df.index, plat_df['Netflix'], marker = 'o',label='Netflix')
plt.plot(plat_df.index, plat_df['Prime Video'], marker = 'x',label='Prime Video')
plt.plot(plat_df.index, plat_df['Hulu'], marker = '*',label='Hulu')
plt.plot(plat_df.index, plat_df['Disney+'], marker = 'X',label='Disney+');
plt.xlabel('Year')
plt.ylabel('No of Shows')
leg = plt.legend();
plt.show()
```

### Which year did all of these platforms have the most shows released?

```python
plat_array=plat_df.to_numpy()
plat_array
```

```python
# Finding maximum value of each column
max_column=np.max(plat_array,axis=0)
print(max_column)
# Finding index of maximum value of each column
max_column_index=np.argmax(plat_array,axis=0)
print(max_column_index)
print('Maximum shows released on each platform are:\n')
print('Netflix :{} shows ,Year:{}'.format(max_column[0],plat_df.index[18]))
print('Prime Video :{} shows ,Year:{}'.format(max_column[1],plat_df.index[17]))
print('Hulu :{} shows ,Year:{}'.format(max_column[2],plat_df.index[18]))
print('Disney+ :{} shows ,Year:{}'.format(max_column[3],plat_df.index[19]))

```

```python
plt.figure(figsize = (14,7))
plt.title('Shows Available on each Platform with their Year of Release')
sns.heatmap(plat_df, linewidths=1, fmt = "d", cmap ="coolwarm")
plt.xlabel('Streaming Platforms');
```

### Is there any change in the number of shows being released from 2019,2020?

```python
release_2020=df[df['Year']>2019]
No_shows_2020=release_2020.Title.nunique()
No_shows_2020
```

```python
Total_shows=df['Title'].nunique()

Total_shows
```

```python
shows_2019=No_shows_2019/Total_shows
shows_2020=No_shows_2020/Total_shows
fall_percentage=(shows_2019-shows_2020)*100
print(fall_percentage)
```
